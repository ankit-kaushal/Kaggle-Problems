### Handling Missing Values

```python
df.info()
```

```python
df.describe()
```

```python
#missing values in each column
df.isnull().sum()
```

### Missing values in 'cabin' column

```python
# this column is MCAR (missing value  completely at random) as it does not affect 'Survived' (target variable) 
#more than 50% values of cabin column is missing so we are creating a new value 'Missing' which we assign to the all the missing values.
df['Cabin'].fillna('Missing',inplace=True)
```

```python
df.head()
```

```python
df['Cabin']=df['Cabin'].astype(str)
df['Cabin']=df['Cabin'].str[0]
```

```python
df.head()
```

### Missing values in 'Emabarked' column

```python
df['Embarked'].unique()
```

```python
df['Embarked'].value_counts()
```

```python
# this column is also MCAR. so we jsut replacing the misisng values with mode of the column 'Embarked'
df['Embarked'].fillna(df['Embarked'].mode()[0],inplace=True)
```

### Missing values in 'Age' column

```python
sns.jointplot('Age','Survived',data=df,kind='hex')
```

```python
#this graph depicts that there is no suc relation between 'Age' and 'Survived'
#so this column is also MCAR as it does not affect survived.
# as it is normally distributed we can replace missing values with median
sns.distplot(df['Age'])
```

```python
df['Age'].fillna(df['Age'].median(),inplace=True)
```

### All the missing values are imputed

```python
df.isnull().sum()
```

### Handling Categorical Features

### Handling 'Sex' column

```python
#as this have only two categoties we can apply one hot encoding over it
df['Sex'].unique()
```

```python
df = pd.get_dummies(df,columns=['Sex'],drop_first=True)
```

```python
df.head()
```

### Handling 'Cabin' column

```python
#one hot encoding
df['Cabin'].unique()
```

```python
df = pd.get_dummies(df,columns=['Cabin'],drop_first=True)
```

```python
df.head()
```

### Handling 'Embarked' column

```python
df['Embarked'].unique()
```

```python
df = pd.get_dummies(data=df,columns=['Embarked'],drop_first=True)
```

```python
df.head()
```

### Train Test Split

```python
X=df.drop('Survived',axis=1)
y=df['Survived']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,stratify=y,random_state=0,test_size=0.2)
```

### Model Creation

```python
#as we are going to use random forest classifier so need need to scale the data
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import RandomizedSearchCV
n_estimators = [int(x) for x in np.linspace(start=200,stop=2000,num=10)]
max_features = ['auto','log2']
max_depth = [int(x) for x in np.linspace(10,100,2)]
min_samples_leaf = [1,2,4,6,8,10,15,20]
random_grid = {'n_estimators':n_estimators,
               'max_features':max_features,
                'max_depth':max_depth,
                 'min_samples_leaf':min_samples_leaf}
random_forest = RandomForestClassifier()
rf_random_cv = RandomizedSearchCV(estimator=random_forest,param_distributions=random_grid,n_iter=100,cv=3,n_jobs=-1)
rf_random_cv.fit(X_train,y_train)
```

```python
rf_random_cv.best_score_
```

```python
rf_random_cv.best_params_
```

```python
rf_random_cv.best_estimator_
```

```python
rf_random_cv.score(X_test,y_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
y_pred=rf_random_cv.predict(X_test)
print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

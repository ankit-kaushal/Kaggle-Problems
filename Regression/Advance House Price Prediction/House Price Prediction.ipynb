### Making all the categories declared as rare whose count is less than 1% in a categorical feature

```python
for i in categorical_cols:
    temp = df.groupby(i)['SalePrice'].count()/len(df)
    temp_df = temp[temp>0.01].index
    df[i]=np.where(df[i].isin(temp_df),df[i],'Rar_var')
```

```python
df.head(31)
```

### converting 'categorical variable' to 'numerical variable'

```python
for i in categorical_cols:
    lp = df.groupby(i)['SalePrice'].mean().sort_values().to_dict()
    ordinal_labels = {k:i for i,k in enumerate(lp,0)}
    df[i]=df[i].map(ordinal_labels)
```

```python
df.head()
```

### Removing features which are highly correlated to each other

```python
X= df.drop('SalePrice',axis=1)
y= df['SalePrice']
```

```python
plt.figure(figsize=(30,20))
sns.heatmap(df.corr(),annot=True)
```

```python
list_columns= []
df_corr = X.corr()
for i in range(len(X.corr().columns)):
    for j in range(i):
        if(abs(X.corr().iloc[i,j])>0.75):
          list_columns.append(df_corr.columns[i])
```

```python
list_columns
```

```python
X.drop(list_columns,axis=1,inplace=True)
```

```python
X
```

### Train Test split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=0,test_size=0.2)
```

### Model Creation

```python
# no scaling is required as we are going to use random forest regressor
from sklearn.model_selection import RandomizedSearchCV
from sklearn.ensemble import RandomForestRegressor
params = { 'n_estimators':[int(x) for x in np.linspace(100,1000,10)],
           'min_samples_leaf':[1,2,3,5,10,15,18,24,32,64],
         'max_depth':[int(x) for x in np.linspace(2,50,30)]}
rf = RandomForestRegressor()
rf_cv = RandomizedSearchCV(estimator=rf,param_distributions=params,n_jobs=-1,verbose=True,cv=5,n_iter=100)
rf_cv.fit(X_train,y_train)
```

```python
rf_cv.best_score_
```

```python
rf_cv.score(X_test,y_test)
```

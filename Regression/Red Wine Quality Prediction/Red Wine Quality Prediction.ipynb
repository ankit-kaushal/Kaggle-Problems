```python
X = df.drop('quality',axis=1)
y = df['quality']
```

### Removing outliers

```python
for i in X.columns:
    upper_boundary = X[i].mean()+3*X[i].std()
    lower_boundary = X[i].mean()-3*X[i].std()
    median = X.loc[X[i]<=upper_boundary,i].median()
    print(median)
    X.loc[X[i]>upper_boundary,i]=median
    X.loc[X[i]<lower_boundary,i]=median
```

### Scaling the features

```python
# as all the features are nearly normally distributes so we applyingng standardization
count=1
plt.figure(figsize=(20,15))
for i in X.columns:
    plt.subplot(3,4,count)
    sns.distplot(df[i])
    count=count+1
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
st = StandardScaler()
st.fit(X_train)
```

```python
X_train = pd.DataFrame(st.transform(X_train),columns=X.columns)
```

```python
X_test = pd.DataFrame(st.transform(X_test),columns=X.columns)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import RandomizedSearchCV
params ={'n_estimators': [int(x) for x in np.linspace(100,1000,10)],
         'max_depth':[1,5,10,15,20,25,30],
         'min_samples_leaf':[1,3,5,8,10]}
rf = RandomForestRegressor()
rf_cv = RandomizedSearchCV(estimator=rf,param_distributions=params,n_iter=50,cv=3,n_jobs=-1,verbose=True)
rf_cv.fit(X_train,y_train)
```
